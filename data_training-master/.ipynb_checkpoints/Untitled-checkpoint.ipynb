```python
import pandas as pd
world = pd.read_csv('Tennis_World_Number_1s.csv')
world
```

```python
world.drop('Unnamed: 7', axis=1, inplace=True)
```

```python
world[['wins','losses']] = world.Careers_Wins_Losses.str.split('-',expand=True).astype(float)
```

```python
world
```

```python
world.dtypes
```

```python
world['Total_Professional_matches'] = world.wins + world.losses
world['Win_Percentage'] = (world.wins/world.Total_Professional_matches)*100
world[['Tennis_Player','Win_Percentage']].sort_values('Win_Percentage',ascending=False).head(1)
```

```python
world[['Tennis_Player','Years_End_no_1','Win_Percentage']].sort_values(['Years_End_no_1','Win_Percentage'],ascending=False)
```

```python
Tenish = world.sort_values('Win_Percentage', ascending=False)
Tenish['percen_diff'] = Tenish.Win_Percentage.pct_change()*100
Tenish['percen_diff'] = Tenish.percen_diff.fillna(0)
Tenish[['Tennis_Player','percen_diff']]
```

```python
cond1 = world.Maximum_Consectutive_Weeks_at_no_1>30
cond2 = world.Titles > 80
cond3 = world.Prize_Money < 100000000
world[cond1 & cond2 & cond3].loc[:,['Tennis_Player']]# no of player is shown
```

```python
import matplotlib.pyplot as plt
#quick inspection(shown)of the data
world.plot.barh(x='Tennis_Player',y='Total_Weeks_at_No_1',figsize=(15,10))
# Filtering the output,Weeks at no 1 greater than 50 will only be shown,
world[world.Total_Weeks_at_No_1 > 50].plot.barh(x='Tennis_Player',y='Total_Weeks_at_No_1',figsize=(15,10))
```

```python
world['Earnings'] = Tenish.Prize_Money/10**6
tenish_players = list(world.Tennis_Player)
x_cords = list(world.Earnings)
y_cords = list(world.Win_Percentage)
plt.figure(figsize=(18,12))
for i, player in  enumerate(tenish_players):
    x = x_cords[i]
    y = y_cords[i]
    plt.scatter(x,y,color='red',marker='o',alpha=0.5,s=100)
    plt.text(x+0.3,y+0.3,player,fontsize=10)
plt.title('The relationship between Win percentage and prize money for former world number no 1s',fontsize=25)    
plt.ylabel('Win Percentage(%)',fontsize=25)
plt.xlabel('Prize Money', fontsize=25)
plt.show()
```
